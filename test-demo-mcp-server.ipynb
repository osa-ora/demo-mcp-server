```python
from fastmcp import Client

client = Client("http://127.0.0.1:8085/mcp")

async with client as c:

    demos = await c.call_tool("list_demos")
    print("DEMOS:", demos)
    demo = await c.call_tool(
        "get_demo",
        {"name": "MCP Server Client Orchestrator"}
    )
    print("DEMO:", demo)
    demo = await c.call_tool(
        "install_demo",
        {"name": "MCP Server Client Orchestrator","workspace": "/Users/ooransa/Desktop/demos/"}
    )
    print("DEMO:", demo)
    demo = await c.call_tool(
        "run_demo",
        {"name": "MCP Server Client Orchestrator","workspace": "/Users/ooransa/Desktop/demos/"}
    )
    print(">>>>>>>DEMO:", demo)
```
